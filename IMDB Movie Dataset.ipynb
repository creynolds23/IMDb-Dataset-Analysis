20. Classify Movies Based on Ratings

```python
data.columns
```

```python
# Create Funtion to classify movies based on ratings
def rating(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating >= 6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Rating_class']=data['Rating'].apply(rating)
```

22. Find Unique Values from Genre

```python
# Get the Genre column, split values, and create a list of unique genres
data['Genre']
```

```python
list1 =[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d = []
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
unique_list = []
for item in one_d:
    if item not in unique_list:
        unique_list.append(item)
        
```

```python
unique_list
```

25. Box Office Success by Genre

```python
# Split the genre data and remove duplicates and create barplot for visual
data['Genre'] = data['Genre'].str.split(',')
data = data.explode('Genre')

plt.figure(figsize=(12,6))
sns.barplot(x='Genre', y='Revenue (Millions)', data=data, errorbar=None)
plt.title("Box Office Success by Genre")
plt.xticks(rotation=90)
plt.show()
```
